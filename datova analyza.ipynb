## Přejmenování sloupce *#* na *Number*

```python
data.rename(columns={'#':'Number'}, inplace=True)
data.head()
```

## Vybereme si jen běžné pokémony

```python
bezni_pokemoni = data[data['Legendary'] == False]
bezni_pokemoni['Legendary'].sum()
```

## Bez všech těch *Mega* nesmyslů

```python
ne_mega = bezni_pokemoni[~bezni_pokemoni['Name'].str.contains('Mega')]
ne_mega.head()
```

## A co další duplicity v datasetu?

```python
ne_mega['Number'].groupby(ne_mega['Number']).count().sort_values(ascending=False).head(10)
```

```python
ne_mega['Number'].count()
```

```python
bez_duplicit = ne_mega.drop_duplicates('Number', keep='first')
bez_duplicit['Number'].count()
```

## Reset indexu v novém datasetu

```python
bez_duplicit.reset_index(inplace=True, drop=True)
bez_duplicit
```

## Které typy pokémonů jsou nejsilnější?

```python
nejsilnejsi_typy = bez_duplicit[['Type 1', 'Total']].groupby(bez_duplicit['Type 1']).mean().sort_values(by='Total', ascending=False)
```

```python
nejsilnejsi_typy
```

## A které nejfrekventovanější?

```python
frekvence_typu = bez_duplicit['Type 1'].groupby(bez_duplicit['Type 1']).count().sort_values(ascending=False)
frekvence_typu
```

## Top 5 nejlepších pokémonů

```python
top5 = bez_duplicit.sort_values(by='Total', ascending=False)[:5]
top5
```

## A co jednotlivé generace?

```python
generace = bez_duplicit[['Generation', 'Total']].groupby(bez_duplicit['Generation']).mean().sort_values(by='Total', ascending=False)
generace
```

```python
generace.plot.bar(x='Generation', y='Total')
```

```python
generace_serazeno = generace.sort_values(by='Generation')
```

```python
generace_serazeno.plot.bar(x='Generation', y='Total')
```

## A zase ten Pikachu … Jak je na tom mezi svými?

```python
pikachu_rank = bez_duplicit[bez_duplicit['Type 1'] == 'Electric'].sort_values(by='Total')
pikachu_rank
```

```python
pikachu_rank.reset_index(inplace=True)
```

```python
pikachu_rank
```

```python
pikachu_rank['color'] = 'Grey'
pikachu_rank['color'][pikachu_rank['Name'] == 'Pikachu'] = 'Red'
pikachu_rank.plot.scatter(x='Generation', y='Total', color=pikachu_rank['color'])
```
