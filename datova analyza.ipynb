## Úkol č. 2 - základní graf

## Přejmenování sloupce *#* na *Number*

```python
data.rename(columns={'#':'Number'}, inplace=True)
data.head()
```

## Vybereme si jen běžné pokémony

```python
bezni_pokemoni = data[data['Legendary'] == False]
bezni_pokemoni['Legendary'].sum()
```

## Bez všech těch *Mega* nesmyslů

```python
bezni_pokemoni[bezni_pokemoni['Name'].str.contains('Mega')]
```

![Meganium](https://proxy.duckduckgo.com/iu/?u=http%3A%2F%2Fimg2.wikia.nocookie.net%2F__cb20140315173207%2Fpokemon%2Fimages%2Fthumb%2F1%2F11%2F154Meganium_OS_anime.png%2F500px-154Meganium_OS_anime.png&f=1)

```python
bezni_pokemoni[bezni_pokemoni['Name'].str.contains(' ') & bezni_pokemoni['Name'].str.contains('Mega')]
```

```python
ne_mega = bezni_pokemoni[~(bezni_pokemoni['Name'].str.contains(' ') & bezni_pokemoni['Name'].str.contains('Mega'))]
ne_mega.head()
```

## A co další duplicity v datasetu?

```python
ne_mega['Number'].groupby(ne_mega['Number']).count().sort_values(ascending=False).head(10)
```

```python
ne_mega[ne_mega['Number'] == 479]
```

```python
ne_mega['Number'].count()
```

```python
bez_duplicit = ne_mega.drop_duplicates('Number', keep='first')
bez_duplicit['Number'].count()
```

## Reset indexu v novém datasetu

```python
bez_duplicit.reset_index(inplace=True, drop=True)
bez_duplicit
```

## Které typy pokémonů jsou nejsilnější?

```python
nejsilnejsi_typy = bez_duplicit.groupby('Type 1').mean().sort_values(by='Total', ascending=False)
```

```python
nejsilnejsi_typy
```

```python
nejsilnejsi_typy.plot(kind='bar', y='Total')
```

## A které nejfrekventovanější?

```python
frekvence_typu = bez_duplicit.groupby('Type 1').size().sort_values(ascending=False)
frekvence_typu
```

```python
frekvence_typu.plot(kind='bar')
```

## Jaké kombinace typů jsou vlastně možné

```python
kombinace_typu = bez_duplicit.groupby(['Type 1', 'Type 2']).size().sort_values(ascending=False)[:20]
kombinace_typu
```

## A co jednotlivé generace?

```python
generace = bez_duplicit.groupby('Generation').mean().sort_values(by='Total', ascending=False)
generace
```

```python
generace.plot.bar(y='Total')
```

```python
generace.sort_index().plot.bar(y='Total')
```

## Úkol č. 3 - skupiny pokémonů a jejich statistiky

## A zase ten Pikachu … Jak je na tom mezi svými?

```python
pikachu_rank = bez_duplicit[bez_duplicit['Type 1'] == 'Electric'].sort_values(by='Total')
pikachu_rank
```

```python
pikachu_rank.reset_index(inplace=True)
```

```python
pikachu_rank
```

```python
pikachu_rank['color'] = 'Grey'
pikachu_rank.loc[pikachu_rank['Name'] == 'Pikachu', 'color'] = 'Red'
pikachu_rank.plot.scatter(x='Generation', y='Total', color=pikachu_rank['color'])
```
